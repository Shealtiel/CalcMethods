```python
import numpy as np
import math
import pandas as pd
from IPython import display
from bokeh.io import output_notebook, show, export_png
from bokeh.plotting import figure
from bokeh.layouts import gridplot
output_notebook()
```

Определяем переменные

```python
a = 0
b = 1
c = 0
d = 10

xlin = 10
tlin = 2500

dbig = 1

xstep = (b - a) / xlin
tstep = (d - c) / tlin

alpha = (dbig ** 2) * tstep / (xstep ** 2)
if 2 * alpha > 1:
    print('Нестабильная система')
    print(alpha)

u = np.zeros((tlin+1, xlin+1))
x = np.linspace(a, b, xlin+1)
t = np.linspace(c, d, tlin+1)

u[0] = 1 + x
u[:, 0] = 1
u[:, xlin] = 2

w  = np.copy(u)
```

Явный метод

```python
for i in range(0, tlin):
    for j in range(1, xlin):
        u[i+1, j] = u[i, j] + alpha * (u[i, j-1] - 2 * u[i, j] + u[i, j+1])
# pd.DataFrame(u)
```

Неявный метод

```python
arr = np.zeros(xlin+1)
matr = np.zeros((xlin+1, xlin+1))
malpha = - alpha

for j in range(1, xlin+1):
        matr[j-1, j] = malpha
        matr[j, j] = 1 - 2 * malpha
        matr[j, j-1] = malpha

matr[0, 0] = 1
matr[0, 1] = 0
matr[xlin, xlin] = 1
matr[xlin, xlin-1] = 0

for i in range(0, tlin):
    arr = w[i]
    w[i+1] = np.linalg.solve(matr, arr)
# pd.DataFrame(w)
```

```python
plots = np.array([])
for i in range(0, tlin+1, 71):
    p = figure(plot_width=200, plot_height=200, title=f'{i*d/tlin}')
    p.line(x, u[i], line_width=1)
#     p.line(x, v[i], line_width=1, line_color='red')
    p.line(x, w[i], line_width=1, line_color='black')

    plots = np.append(plots, [p])

pl = []
plots = plots.reshape(6, 6)
for i in plots:
    pl.append(i)

g = gridplot(pl)
# export_png(g, filename="plots_parabolic.png")
show(g)
```
