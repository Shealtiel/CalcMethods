```python
import numpy as np
import pandas as pd
```

Определяем функции

```python
def U(x):
    if x <= 1/2: 
        return 1
    else: 
        return 0


def f(x):
    return x * x
```

Объявляем переменные и устанавливаем начальное условие

```python
h = .1
tau = .1
x = 0
a = np.zeros((12,12))

for i in range(0, 11):
    q = x + i * h
    a[0][i] = U(q)
```

Трехточечная явная схема

```python
for i in range(0, 11):
    for j in range(0, 11):
        next_x = x + i * h
        a[i+1][j] = a[i][j] - next_x * tau / h * (a[i][j] - a[i][j-1]) \
        + tau * f(next_x)
```

Четырехточечная неявная схема. Следует перезапустить блок с переменными для обнуления значений.

```python
for i in range(0, 11):
    for j in range(0, 11):
        next_x = x + i * h
        a[i+1][j] = a[i][j] + next_x * tau / (2 * h) * (a[i+1][j+1] - a[i+1][j-1]) \
        + tau * f(x + (i+1) * h)
```

Импортируем в датафрейм pandas для управления данными

```python
df = pd.DataFrame(a)
pd.set_option('expand_frame_repr', True)
dfr = df.round(3)
dfr.loc[:10, :10]
```
